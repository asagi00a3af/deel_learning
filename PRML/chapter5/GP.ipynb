```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x, beta=0):
    y = np.sin(x*2*np.pi) #* np.exp(-(x**2) / (2*0.4**2))
    n = np.random.normal(0, beta, size=y.shape)
    return y + n

_x = np.arange(-1, 1, 0.01)
_y = f(_x)
plt.plot(_x, _y)
```
